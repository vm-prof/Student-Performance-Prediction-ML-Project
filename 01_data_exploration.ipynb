```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("student_learning_style.csv")

print("Dataset Shape:", df.shape)
print(df.head())
print(df.info())
print(df.columns)

# Visualization
# Gender vs Performance
plt.figure(figsize=(6,4))
sns.boxplot(x="gender", y="math score", data=df)
plt.title("Math Score by Gender")
plt.show()
# Lunch vs Test Scores
sns.barplot(x="lunch", y="reading score", data=df)
plt.title("Reading Score by Lunch Type")
plt.show()
# Test Preparation vs Writing Score
sns.barplot(x="test preparation course", y="writing score", data=df)
plt.title("Writing Score vs Test Preparation")
plt.show()
# creating new features
df['total_score'] = df['math score'] + df['reading score'] + df['writing score']
df['average_score'] = df['total_score'] / 3

df[['math score','reading score','writing score','total_score','average_score']].head()
# Visualization & Insights
plt.figure(figsize=(6,4))
sns.barplot(x='gender', y='average_score', data=df)
plt.title('Average Score by Gender')
plt.show()
# Parental Education vs Average Score
plt.figure(figsize=(10,4))
sns.barplot(x='parental level of education', y='average_score', data=df)
plt.title('Parent Education Impact on Scores')
plt.xticks(rotation=45)
plt.show()
# Test Prep Course vs Average Score
plt.figure(figsize=(6,4))
sns.barplot(x='test preparation course', y='average_score', data=df)
plt.title('Test Preparation Impact')
plt.show()
# Heatmap (Numeric Features Correlation)
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

```

### Insights from Visualizations
- Students who completed **test preparation** scored significantly higher.
- Higher **parental education level**  higher student scores.
- Slight variation between male & female scores in different subjects.
- Math, reading & writing scores are strongly positively correlated.
